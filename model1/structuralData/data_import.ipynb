```python
naio_82x13 = pd.read_excel('naio_82x13.xlsx', sheet_name = 'Sheet1')
#pd.set_option('display.max_rows', 1000)
naio_82x13 = naio_82x13.iloc[9:93]
naio_82x13 = naio_82x13.reset_index(drop=True)
naio_82x13 = naio_82x13.dropna(axis=1, how='all')
naio_82x13 = naio_82x13.loc[:, (naio_82x13 != ":").any()]
naio_82x13.columns=naio_82x13.iloc[0,:]
naio_82x13=naio_82x13.drop(index=[0,1]) #empty values for all the row
naio_82x13 = naio_82x13.reset_index(drop=True)


#use table at basic prices [naio_10_cp1610__custom_13832196]
naio_82x13
```

```python
sbs = pd.read_excel('sbs.xlsx', sheet_name = 'Sheet 1')
#pd.set_option('display.max_rows', 1000)
#sbs = sbs.drop(index=9) #empty values for all the row
sbs = sbs.iloc[8:26]
sbs = sbs.reset_index(drop=True)
sbs = sbs.replace(":", np.nan)
sbs = sbs.replace("e", np.nan) #e = estimated
sbs = sbs.replace("c", np.nan) #c = confidential
sbs.columns=sbs.iloc[0,:]
sbs = sbs.dropna(axis=1,how='all')
sbs = sbs.drop(index=[0,1]) #empty values for all the row
sbs = sbs.reset_index(drop=True)


#here each cell represents the number of people employed
sbs
```
