## media geometrica

```python
import math

def geometric_ratio_from_sum(a1: float, n: int, S: float, tol: float = 1e-12, max_iter: int = 200) -> float:
    """
    Solve for q in (0,1) such that:
        S = a1 * (1 - q**n) / (1 - q)     (q != 1)
    assuming a decreasing geometric series (0<q<1) and a1>0.

    Uses bisection (guaranteed convergence if inputs are consistent).
    """
    if n <= 0:
        raise ValueError("n must be a positive integer.")
    if a1 <= 0:
        raise ValueError("a1 must be > 0 for the decreasing-series assumption.")
    if S <= 0:
        raise ValueError("S must be > 0.")

    # Consistency checks for 0<q<1:
    # As q -> 0+, sum -> a1
    # As q -> 1-, sum -> n*a1
    if S < a1 - 1e-15:
        raise ValueError("Inconsistent data: for 0<q<1, S must be >= a1.")
    if S > n * a1 + 1e-15:
        raise ValueError("Inconsistent data: for 0<q<1, S must be <= n*a1.")

    # Handle edge cases near q=0 or q=1
    if abs(S - a1) <= tol:
        return 0.0
    if abs(S - n * a1) <= tol:
        return 1.0  # limit case (not strictly decreasing, but the limiting value)

    def sum_q(q: float) -> float:
        # Stable evaluation near q ~ 1 via fallback to limit
        if abs(1.0 - q) < 1e-15:
            return n * a1
        return a1 * (1.0 - q**n) / (1.0 - q)

    def f(q: float) -> float:
        return sum_q(q) - S

    lo, hi = 0.0, 1.0
    flo, fhi = f(lo), f(hi - 1e-15)  # avoid exact 1 in computation
    # flo should be <= 0 and fhi >= 0 for consistent inputs
    if flo > 0 or fhi < 0:
        raise ValueError("Could not bracket the root in (0,1). Check inputs.")

    for _ in range(max_iter):
        mid = (lo + hi) / 2.0
        fmid = f(mid)
        if abs(fmid) <= tol or (hi - lo) <= tol:
            return mid
        if fmid > 0:
            hi = mid
        else:
            lo = mid

    raise RuntimeError("Bisection did not converge within max_iter.")

```

## modificazione serie geometrica

Vorrei operare su una serie decrescente di cui conosco primo termine, numero dei termini e totale dei termini. Usiamo lo schema delle progressioni geometriche, ma aggiungendo un correzione che renda variabile la ragione, sempre < 1, ma via via crescente in modo tale che la serie scenda rapidamente ma poi si stabilizzi; mi serve poter indicare il punto in cui inizia la stabilizzazione.

Di seguito trovi uno schema “alla progressione geometrica” con **ragione variabile** $r_i$, sempre $<1$, **crescente** (quindi la serie decresce rapidamente all’inizio e poi si stabilizza). Tu indichi:

* `a1` = primo termine $a_1$
* `n` = numero termini
* `S` = somma totale $\sum_{i=1}^n a_i$
* `i0` = **punto di stabilizzazione** (indice del termine da cui la ragione si avvicina a $r_{\max}$ e la discesa rallenta)

La ragione è costruita come transizione logistica:

$$
r_i = r_{\min} + (r_{\max}-r_{\min})\cdot \sigma\left(\frac{i - i_0}{\tau}\right),
\quad \sigma(x)=\frac{1}{1+e^{-x}},
\quad i=1,\dots,n-1
$$

con $0<r_{\min}<r_{\max}<1$ e $\tau>0$ che controlla “quanto graduale” è la stabilizzazione.

$\textcolor{red}{\text{Qui sotto}}$ il **codice Python completo** (solo `numpy`) che **calibra automaticamente (r_{\min})** via bisezione per far tornare esattamente la somma $S$, dati `a1, n, S, i0` (e scelte ragionevoli per `r_max, tau`).


### Come “pilotare” la stabilizzazione (in pratica)

* `i0` (intero o anche float) sposta **il punto** in cui la ragione “sale” verso `r_max`.

  * Più **piccolo** → stabilizzazione prima.
  * Più **grande** → discesa rapida più lunga.
* `tau` controlla **la gradualità**:

  * piccolo (es. 1–2) → “ginocchio” netto,
  * grande (es. 6–10) → transizione molto morbida.
* `r_max` quanto “piatta” diventa la coda (sempre < 1).

Se mi dai un tuo terzetto reale (`a1`, `n`, `S`) e un’idea di dove vuoi il “ginocchio” (`i0`), posso suggerirti valori iniziali sensati di `tau` e `r_max` per ottenere una forma più vicina a quella che hai in mente.



```python
import numpy as np

def _sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def _build_ratios(n, i0, tau, r_min, r_max):
    """
    Costruisce r_i per i=1..n-1 (array di lunghezza n-1), crescente e < 1.
    i0: indice (1-based) del punto attorno a cui la transizione accelera verso r_max.
    """
    i = np.arange(1, n)  # i = 1..n-1
    # Transizione centrata su i0: quando i ~ i0 la logistica è ~0.5
    x = (i - i0) / tau
    return r_min + (r_max - r_min) * _sigmoid(x)

def _series_from_ratios(a1, ratios):
    """
    Dato a1 e ratios r_1..r_{n-1}, produce a_1..a_n con a_{k+1}=a_k*r_k.
    """
    n = len(ratios) + 1
    a = np.empty(n, dtype=float)
    a[0] = float(a1)
    for k in range(n - 1):
        a[k + 1] = a[k] * ratios[k]
    return a

def generate_stabilizing_geometric_series(
    a1, n, S,
    i0,
    tau=5.0,
    r_max=0.999,
    tol=1e-12,
    max_iter=200
):
    """
    Genera una serie decrescente a_1..a_n con ragione variabile r_i crescente (<1),
    calibrando r_min affinché sum(a)=S, dati a1,n,S,i0.

    Parametri:
      a1    : primo termine (positivo)
      n     : numero termini (>=2)
      S     : somma totale desiderata (>= a1, tipicamente)
      i0    : indice (1-based) del "punto di stabilizzazione" (tra 1 e n-1 consigliato)
      tau   : larghezza transizione (più grande = più graduale)
      r_max : limite superiore della ragione (deve essere < 1)
      tol   : tolleranza su somma
      max_iter: iterazioni bisezione

    Ritorna:
      a      : array (n,) dei termini
      ratios : array (n-1,) delle ragioni r_1..r_{n-1}
      r_min  : valore calibrato di r_min
    """
    a1 = float(a1)
    S = float(S)
    n = int(n)
    i0 = float(i0)
    tau = float(tau)
    r_max = float(r_max)

    if n < 2:
        raise ValueError("n deve essere >= 2.")
    if a1 <= 0:
        raise ValueError("a1 deve essere > 0 (serie decrescente positiva).")
    if not (0.0 < r_max < 1.0):
        raise ValueError("r_max deve essere in (0,1).")
    if tau <= 0:
        raise ValueError("tau deve essere > 0.")
    if S < a1:
        raise ValueError("S deve essere >= a1, perché la somma include almeno il primo termine.")

    # Funzione obiettivo: f(r_min) = sum(a(r_min)) - S
    def f(r_min):
        ratios = _build_ratios(n=n, i0=i0, tau=tau, r_min=r_min, r_max=r_max)
        a = _series_from_ratios(a1, ratios)
        return a.sum() - S

    # Intervallo per r_min: (eps, r_max-eps)
    eps = 1e-12
    lo = eps
    hi = r_max - eps
    if hi <= lo:
        raise ValueError("Intervallo non valido: r_max troppo vicino a 0 o 1.")

    flo = f(lo)
    fhi = f(hi)

    # Monotonia: all'aumentare di r_min, la serie decresce meno rapidamente => somma cresce => f cresce.
    # Serve quindi che S sia tra le somme ottenute ai due estremi.
    if flo > 0:
        # Anche con r_min quasi 0 la somma supera S => S troppo piccolo rispetto a a1 e alla forma.
        raise ValueError(
            "Somma S troppo piccola per i vincoli dati (a1, n, i0, tau, r_max). "
            "Prova a ridurre r_max o aumentare n."
        )
    if fhi < 0:
        # Anche con r_min vicino a r_max la somma resta sotto S => S troppo grande.
        raise ValueError(
            "Somma S troppo grande per i vincoli dati (a1, n, i0, tau, r_max). "
            "Prova ad aumentare r_max (restando <1) o aumentare tau / spostare i0."
        )

    # Bisezione
    for _ in range(max_iter):
        mid = 0.5 * (lo + hi)
        fmid = f(mid)
        if abs(fmid) <= tol:
            r_min_star = mid
            break
        if fmid > 0:
            hi = mid
        else:
            lo = mid
    else:
        r_min_star = 0.5 * (lo + hi)

    ratios = _build_ratios(n=n, i0=i0, tau=tau, r_min=r_min_star, r_max=r_max)
    a = _series_from_ratios(a1, ratios)

    # Piccola correzione numerica: garantiamo somma entro tolleranza (senza alterare a1).
    # Non forzo alcuna riscalatura perché a1 è vincolato.
    if abs(a.sum() - S) > 1e-8:
        # Se vuoi essere più permissivo, puoi alzare max_iter o ridurre tol.
        pass

    return a, ratios, r_min_star


# --- ESEMPIO D'USO ---
if __name__ == "__main__":
    a1 = 100.0
    n = 40
    S = 550.0

    i0 = 10      # da ~qui inizia la stabilizzazione
    tau = 3.0    # quanto "dolce" la transizione
    r_max = 0.995

    a, r, r_min = generate_stabilizing_geometric_series(a1, n, S, i0, tau=tau, r_max=r_max)

    print("r_min calibrato =", r_min)
    print("Somma =", a.sum())
    print("Primi 12 termini:", np.round(a[:12], 6))
    print("Ultimi 5 termini:", np.round(a[-5:], 6))
```
