```python
def total_employees(firms_by_size, n_start=1):
    return sum((n_start + i) * count for i, count in enumerate(firms_by_size))


def shift_firms_to_adjust_total(firms_by_size, delta_employees, n_start=1):
    """
    Sposta imprese tra classi dimensionali adiacenti (passo 1 dipendente)
    per ottenere un aumento/diminuzione del totale dipendenti.

    - Il numero totale di imprese resta costante (si fanno solo spostamenti).
    - Il risultato è approssimato: si ferma quando:
        a) raggiunge delta_employees, oppure
        b) non sono più possibili spostamenti nella direzione richiesta.

    firms_by_size[i] = # imprese con (n_start + i) dipendenti

    Returns:
        new_firms_by_size, achieved_delta, info_dict
    """
    f = list(firms_by_size)
    N = len(f)
    if N <= 1 or delta_employees == 0:
        return f, 0, {"steps": 0, "requested_delta": delta_employees, "achieved_delta": 0}

    requested = int(delta_employees)
    achieved = 0
    steps = 0

    if requested > 0:
        # Per aumentare L, sposto imprese verso destra: i -> i+1 (incremento +1 per impresa spostata)
        remaining = requested
        for i in range(N - 1):  # non posso spostare dall'ultimo verso destra
            if remaining <= 0:
                break
            movable = min(f[i], remaining)  # quante imprese posso spostare da i a i+1
            if movable > 0:
                f[i] -= movable
                f[i + 1] += movable
                achieved += movable
                remaining -= movable
                steps += movable

    else:
        # Per diminuire L, sposto imprese verso sinistra: i -> i-1 (decremento -1 per impresa spostata)
        remaining = -requested
        for i in range(N - 1, 0, -1):  # non posso spostare dal primo verso sinistra
            if remaining <= 0:
                break
            movable = min(f[i], remaining)
            if movable > 0:
                f[i] -= movable
                f[i - 1] += movable
                achieved -= movable
                remaining -= movable
                steps += movable

    info = {
        "steps": steps,
        "requested_delta": requested,
        "achieved_delta": achieved,
        "total_firms_before": sum(firms_by_size),
        "total_firms_after": sum(f),
        "total_employees_before": total_employees(firms_by_size, n_start=n_start),
        "total_employees_after": total_employees(f, n_start=n_start),
    }
    return f, achieved, info
```

```python
import heapq

def total_employees(firms_by_size, n_start=1):
    return sum((n_start + i) * count for i, count in enumerate(firms_by_size))


def _adjust_right_min_l2(f, delta, weights="inv", batch_size=256, max_steps=None):
    """
    Sposta imprese verso destra (i -> i+1) per aumentare il totale dipendenti di circa 'delta' (delta>0),
    minimizzando J = sum_i w_i * (f'_i - f_i)^2.

    f: lista di conteggi (non negativa)
    delta: incremento desiderato (int >= 0)
    weights: "inv" (1/max(f_i,1)) oppure "uniform"
    batch_size: per accelerare, applica mosse in blocchi (approssimazione controllata)
    max_steps: limita il numero di passi effettivi (utile se delta enorme)
    """
    f0 = list(f)
    N = len(f0)
    if delta <= 0 or N <= 1:
        return f0, 0

    # pesi w_i
    if weights == "uniform":
        w = [1.0] * N
    elif weights == "inv":
        w = [1.0 / max(c, 1) for c in f0]
    else:
        raise ValueError("weights must be 'uniform' or 'inv'")

    # m_k: mosse da k a k+1, k = 0..N-2
    m = [0] * (N - 1)

    # deviazioni d_i = f'_i - f_i = m_{i-1} - m_i, con m_{-1}=0, m_{N-1}=0
    d = [0] * N

    def inc_cost(k, t=1):
        """
        Incremento di costo se aumentiamo m_k di t (t>=1).
        Questo cambia solo d_k -= t e d_{k+1} += t.
        """
        dk = d[k]
        dk1 = d[k + 1]
        return (
            w[k] * ((dk - t) ** 2 - dk ** 2) +
            w[k + 1] * ((dk1 + t) ** 2 - dk1 ** 2)
        )

    # heap di (costo marginale per 1, k)
    heap = []
    for k in range(N - 1):
        if f0[k] > 0:  # altrimenti non posso muovere nulla da k
            heapq.heappush(heap, (inc_cost(k, 1), k))

    remaining = int(delta)
    achieved = 0
    steps_done = 0
    if max_steps is None:
        max_steps = remaining

    while remaining > 0 and steps_done < max_steps and heap:
        cost, k = heapq.heappop(heap)

        # se non ho più capacità su k, salto
        cap = f0[k] - m[k]
        if cap <= 0:
            continue

        # blocco di mosse: approssimazione (evita loop enorme quando delta è grande)
        t = min(remaining, cap, batch_size, max_steps - steps_done)
        if t <= 0:
            break

        # applica t mosse su k
        m[k] += t
        d[k] -= t
        d[k + 1] += t

        remaining -= t
        achieved += t
        steps_done += t

        # aggiorna costi locali (k-1, k, k+1) perché d cambia solo su k e k+1
        for kk in (k - 1, k, k + 1):
            if 0 <= kk < N - 1 and (f0[kk] - m[kk] > 0):
                heapq.heappush(heap, (inc_cost(kk, 1), kk))

    # costruisci f' da f0 e m
    f_new = f0[:]
    for k in range(N - 1):
        t = m[k]
        if t:
            f_new[k] -= t
            f_new[k + 1] += t

    return f_new, achieved


def shift_firms_min_distance(firms_by_size, delta_employees, n_start=1,
                            weights="inv", batch_size=256, max_steps=None):
    """
    Versione "min distanza" (L2 pesata): sposta imprese tra classi adiacenti
    per ottenere una variazione del totale dipendenti (delta_employees),
    mantenendo costante il numero totale di imprese.

    Se delta_employees > 0: sposta verso destra (i -> i+1)
    Se delta_employees < 0: sposta verso sinistra (i -> i-1) usando simmetria (reverse)
    """
    f = list(firms_by_size)
    delta = int(delta_employees)

    L0 = total_employees(f, n_start=n_start)
    F0 = sum(f)

    if delta == 0 or len(f) <= 1:
        info = {
            "requested_delta": delta,
            "achieved_delta": 0,
            "total_firms_before": F0,
            "total_firms_after": F0,
            "total_employees_before": L0,
            "total_employees_after": L0
        }
        return f, 0, info

    if delta > 0:
        f_new, achieved = _adjust_right_min_l2(
            f, delta, weights=weights, batch_size=batch_size, max_steps=max_steps
        )
    else:
        # per diminuire: ribalto, aumento, poi ribalto indietro
        f_rev = list(reversed(f))
        f_new_rev, achieved_pos = _adjust_right_min_l2(
            f_rev, -delta, weights=weights, batch_size=batch_size, max_steps=max_steps
        )
        f_new = list(reversed(f_new_rev))
        achieved = -achieved_pos

    L1 = total_employees(f_new, n_start=n_start)
    F1 = sum(f_new)

    info = {
        "requested_delta": delta,
        "achieved_delta": achieved,
        "total_firms_before": F0,
        "total_firms_after": F1,
        "total_employees_before": L0,
        "total_employees_after": L1,
        "weights": weights,
        "batch_size": batch_size,
        "max_steps": max_steps
    }
    return f_new, achieved, info
```

```python
import heapq

def total_employees(firms_by_size, n_start=1):
    return sum((n_start + i) * c for i, c in enumerate(firms_by_size))


def shift_firms_min_distance_monotone(
    firms_by_size,
    delta_employees,
    n_start=1,
    monotone="auto",     # "increasing", "decreasing", "none", "auto"
    weights="inv",       # "inv" (1/max(f0,1)) oppure "uniform"
    max_steps=None
):
    """
    Sposta imprese tra classi adiacenti per ottenere una variazione del totale dipendenti (delta_employees),
    mantenendo:
      - numero totale imprese costante
      - (opzionale) monotonicità della serie f_i (crescente o decrescente)
      - minimizzazione approssimata di una distanza L2 pesata dalla distribuzione originale.

    Nota: se il vincolo di monotonicità impedisce di raggiungere il delta richiesto,
          la funzione si ferma e restituisce achieved_delta < requested_delta.
    """
    f0 = list(firms_by_size)
    f = f0[:]  # stato corrente
    N = len(f)
    if N <= 1:
        return f, 0, {"requested_delta": int(delta_employees), "achieved_delta": 0}

    requested = int(delta_employees)
    if requested == 0:
        return f, 0, {
            "requested_delta": 0,
            "achieved_delta": 0,
            "total_firms_before": sum(f0),
            "total_firms_after": sum(f),
            "total_employees_before": total_employees(f0, n_start=n_start),
            "total_employees_after": total_employees(f, n_start=n_start),
        }

    # determina monotonicità richiesta
    def is_non_decreasing(a):
        return all(a[i] <= a[i+1] for i in range(len(a)-1))
    def is_non_increasing(a):
        return all(a[i] >= a[i+1] for i in range(len(a)-1))

    if monotone == "auto":
        if is_non_decreasing(f0):
            monotone_mode = "increasing"
        elif is_non_increasing(f0):
            monotone_mode = "decreasing"
        else:
            monotone_mode = "none"
    else:
        monotone_mode = monotone

    if monotone_mode not in ("increasing", "decreasing", "none"):
        raise ValueError("monotone must be 'increasing', 'decreasing', 'none', or 'auto'")

    # pesi w_i per J = sum w_i*(f_i - f0_i)^2
    if weights == "uniform":
        w = [1.0] * N
    elif weights == "inv":
        w = [1.0 / max(c, 1) for c in f0]
    else:
        raise ValueError("weights must be 'uniform' or 'inv'")

    # deviazioni correnti d_i
    d = [f[i] - f0[i] for i in range(N)]

    # costo marginale per una mossa a destra su edge k: k -> k+1
    # cambia d_k -= 1 e d_{k+1} += 1
    def inc_cost_right(k):
        dk = d[k]
        dk1 = d[k+1]
        return (
            w[k] * ((dk - 1) ** 2 - dk ** 2) +
            w[k+1] * ((dk1 + 1) ** 2 - dk1 ** 2)
        )

    # costo marginale per una mossa a sinistra su edge k: (k+1) -> k
    # equivalente a aumentare f_k e diminuire f_{k+1}
    # cambia d_k += 1 e d_{k+1} -= 1
    def inc_cost_left(k):
        dk = d[k]
        dk1 = d[k+1]
        return (
            w[k] * ((dk + 1) ** 2 - dk ** 2) +
            w[k+1] * ((dk1 - 1) ** 2 - dk1 ** 2)
        )

    # verifica se una mossa è ammissibile rispetto alla monotonicità
    def feasible_right(k):
        if f[k] <= 0:
            return False
        if monotone_mode == "none":
            return True
        # prova mossa unitaria
        a = f[k] - 1
        b = f[k+1] + 1
        if monotone_mode == "increasing":
            if k-1 >= 0 and f[k-1] > a:      # f[k-1] <= new f[k]
                return False
            if a > b:                         # new f[k] <= new f[k+1]
                return False
            if k+2 < N and b > f[k+2]:        # new f[k+1] <= f[k+2]
                return False
            return True
        else:  # "decreasing"
            if k-1 >= 0 and f[k-1] < a:      # f[k-1] >= new f[k]
                return False
            if a < b:                        # new f[k] >= new f[k+1]
                return False
            if k+2 < N and b < f[k+2]:       # new f[k+1] >= f[k+2]
                return False
            return True

    def feasible_left(k):
        # mossa unitaria (k+1)->k
        if f[k+1] <= 0:
            return False
        if monotone_mode == "none":
            return True
        a = f[k] + 1
        b = f[k+1] - 1
        if monotone_mode == "increasing":
            if k-1 >= 0 and f[k-1] > a:
                return False
            if a > b:
                return False
            if k+2 < N and b > f[k+2]:
                return False
            return True
        else:  # "decreasing"
            if k-1 >= 0 and f[k-1] < a:
                return False
            if a < b:
                return False
            if k+2 < N and b < f[k+2]:
                return False
            return True

    # heap con le mosse candidate: (deltaJ, k)
    heap = []
    direction = "right" if requested > 0 else "left"
    if direction == "right":
        for k in range(N-1):
            if feasible_right(k):
                heapq.heappush(heap, (inc_cost_right(k), k))
    else:
        for k in range(N-1):
            if feasible_left(k):
                heapq.heappush(heap, (inc_cost_left(k), k))

    remaining = abs(requested)
    achieved = 0
    steps = 0
    if max_steps is None:
        max_steps = remaining

    while remaining > 0 and steps < max_steps and heap:
        cost, k = heapq.heappop(heap)

        if direction == "right":
            if not feasible_right(k):
                continue
            # applica la mossa
            f[k] -= 1
            f[k+1] += 1
            d[k] -= 1
            d[k+1] += 1
            achieved += 1
        else:
            if not feasible_left(k):
                continue
            f[k] += 1
            f[k+1] -= 1
            d[k] += 1
            d[k+1] -= 1
            achieved -= 1

        remaining -= 1
        steps += 1

        # aggiorna localmente le mosse che possono cambiare ammissibilità/costo
        for kk in (k-2, k-1, k, k+1):
            if 0 <= kk < N-1:
                if direction == "right":
                    if feasible_right(kk):
                        heapq.heappush(heap, (inc_cost_right(kk), kk))
                else:
                    if feasible_left(kk):
                        heapq.heappush(heap, (inc_cost_left(kk), kk))

    info = {
        "monotone_mode": monotone_mode,
        "direction": direction,
        "requested_delta": requested,
        "achieved_delta": achieved,
        "steps": steps,
        "total_firms_before": sum(f0),
        "total_firms_after": sum(f),
        "total_employees_before": total_employees(f0, n_start=n_start),
        "total_employees_after": total_employees(f, n_start=n_start),
        "weights": weights,
    }
    return f, achieved, info
```

```python
def weighted_average(series,start):
    num=0
    den=0
    for i in range(start,len(series)+start):
        num+=i*series[i-start]
        den+=series[i-start]
    return num/den
```

```python
import heapq
import math

def total_employees(firms_by_size, n_start=1):
    return sum((n_start + i) * c for i, c in enumerate(firms_by_size))


def shift_firms_preserve_ratios_monotone(
    firms_by_size,
    delta_employees,
    n_start=1,
    eps=1.0,
    monotone="auto",      # "auto", "increasing", "decreasing", "none"
    max_steps=None,
    refresh_window=2
):
    """
    Sposta imprese tra classi adiacenti per ottenere delta_employees (approssimato),
    preservando:
      - numero totale imprese (solo spostamenti)
      - monotonicità (se richiesta)
      - rapporti adiacenti f[i+1]/f[i] il più possibile (minimizzazione greedy della variazione
        dei log-rapporti rispetto all'originale)

    Obiettivo: minimizzare sum_i ( log((f'_{i+1}+eps)/(f'_i+eps)) - log((f_{i+1}+eps)/(f_i+eps)) )^2
    """
    f0 = list(firms_by_size)
    f = f0[:]
    N = len(f)

    requested = int(delta_employees)
    if N <= 1 or requested == 0:
        return f, 0, {
            "requested_delta": requested,
            "achieved_delta": 0,
            "monotone_mode": "none",
            "total_firms_before": sum(f0),
            "total_firms_after": sum(f),
            "total_employees_before": total_employees(f0, n_start=n_start),
            "total_employees_after": total_employees(f, n_start=n_start),
        }

    # monotonicità: auto-detect se richiesto
    def is_non_decreasing(a): return all(a[i] <= a[i+1] for i in range(len(a)-1))
    def is_non_increasing(a): return all(a[i] >= a[i+1] for i in range(len(a)-1))

    if monotone == "auto":
        if is_non_decreasing(f0):
            monotone_mode = "increasing"
        elif is_non_increasing(f0):
            monotone_mode = "decreasing"
        else:
            monotone_mode = "none"
    else:
        monotone_mode = monotone

    if monotone_mode not in ("increasing", "decreasing", "none"):
        raise ValueError("monotone must be 'auto', 'increasing', 'decreasing', or 'none'")

    # log-ratio base (target) dalla distribuzione originale
    base_lr = [math.log(f0[i+1] + eps) - math.log(f0[i] + eps) for i in range(N-1)]

    def local_cost_at(i, arr):
        lr = math.log(arr[i+1] + eps) - math.log(arr[i] + eps)
        d = lr - base_lr[i]
        return d * d

    def local_cost_sum(i_from, i_to, arr):
        a = max(0, i_from)
        b = min(N - 2, i_to)
        s = 0.0
        for i in range(a, b + 1):
            s += local_cost_at(i, arr)
        return s

    # monotonic feasibility checks for a unit move on edge k between k and k+1
    def feasible_right(k):
        # move 1 firm: k -> k+1  (f[k]--, f[k+1]++)
        if f[k] <= 0:
            return False
        if monotone_mode == "none":
            return True

        a = f[k] - 1
        b = f[k+1] + 1

        if monotone_mode == "increasing":
            if k-1 >= 0 and f[k-1] > a:      # f[k-1] <= new f[k]
                return False
            if a > b:                         # new f[k] <= new f[k+1]
                return False
            if k+2 < N and b > f[k+2]:        # new f[k+1] <= f[k+2]
                return False
            return True

        # decreasing
        if k-1 >= 0 and f[k-1] < a:          # f[k-1] >= new f[k]
            return False
        if a < b:                             # new f[k] >= new f[k+1]
            return False
        if k+2 < N and b < f[k+2]:            # new f[k+1] >= f[k+2]
            return False
        return True

    def feasible_left(k):
        # move 1 firm: (k+1) -> k  (f[k]++, f[k+1]--)
        if f[k+1] <= 0:
            return False
        if monotone_mode == "none":
            return True

        a = f[k] + 1
        b = f[k+1] - 1

        if monotone_mode == "increasing":
            if k-1 >= 0 and f[k-1] > a:
                return False
            if a > b:
                return False
            if k+2 < N and b > f[k+2]:
                return False
            return True

        # decreasing
        if k-1 >= 0 and f[k-1] < a:
            return False
        if a < b:
            return False
        if k+2 < N and b < f[k+2]:
            return False
        return True

    # marginal cost of a unit move (ratio-based objective), computed locally
    def inc_cost_right(k):
        if not feasible_right(k):
            return float("inf")
        before = local_cost_sum(k - 1, k + 1, f)
        ff = f[:]              # simple but safe; can be optimized if needed
        ff[k] -= 1
        ff[k+1] += 1
        after = local_cost_sum(k - 1, k + 1, ff)
        return after - before

    def inc_cost_left(k):
        if not feasible_left(k):
            return float("inf")
        before = local_cost_sum(k - 1, k + 1, f)
        ff = f[:]
        ff[k] += 1
        ff[k+1] -= 1
        after = local_cost_sum(k - 1, k + 1, ff)
        return after - before

    direction = "right" if requested > 0 else "left"
    remaining = abs(requested)
    achieved = 0
    steps = 0
    if max_steps is None:
        max_steps = remaining

    heap = []
    if direction == "right":
        for k in range(N-1):
            if feasible_right(k):
                heapq.heappush(heap, (inc_cost_right(k), k))
    else:
        for k in range(N-1):
            if feasible_left(k):
                heapq.heappush(heap, (inc_cost_left(k), k))

    while remaining > 0 and steps < max_steps and heap:
        cost, k = heapq.heappop(heap)

        if direction == "right":
            if not feasible_right(k):
                continue
            f[k] -= 1
            f[k+1] += 1
            achieved += 1
        else:
            if not feasible_left(k):
                continue
            f[k] += 1
            f[k+1] -= 1
            achieved -= 1

        remaining -= 1
        steps += 1

        # refresh local neighborhood of candidate moves
        lo = max(0, k - refresh_window)
        hi = min(N - 2, k + refresh_window)
        for kk in range(lo, hi + 1):
            if direction == "right":
                if feasible_right(kk):
                    heapq.heappush(heap, (inc_cost_right(kk), kk))
            else:
                if feasible_left(kk):
                    heapq.heappush(heap, (inc_cost_left(kk), kk))

    info = {
        "requested_delta": requested,
        "achieved_delta": achieved,
        "steps": steps,
        "direction": direction,
        "monotone_mode": monotone_mode,
        "eps": eps,
        "total_firms_before": sum(f0),
        "total_firms_after": sum(f),
        "total_employees_before": total_employees(f0, n_start=n_start),
        "total_employees_after": total_employees(f, n_start=n_start),
    }
    return f, achieved, info
```
