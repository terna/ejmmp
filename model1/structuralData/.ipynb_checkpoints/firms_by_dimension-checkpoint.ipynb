$\qquad  
b_i = b_1 - (i-1)\,\Delta b, \qquad \Delta b > 0$


**Linear case**

With $m$ mean value:  
$\Delta b = \frac{2\,(b_1 - m)}{n-1}, \qquad b_1 > m$

```python
# prova
b_1 =  3
S   = 28
n   = 10
m   = S/n
b_n = 2*S/n - b_1
b_n
```

```python
Db = (b_1 - b_n)/(n-1)
Db
```

```python
ist=[]
for i in range(1,11):
    ist.append(b_1-(i - 1)*Db)
ist
```
