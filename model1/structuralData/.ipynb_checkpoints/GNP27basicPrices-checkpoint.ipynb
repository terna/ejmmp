```python
#in test_ff_multi 
GDP27currentPrices = 16144780 #difference due to update values; contains VAT
GDP27basicPrices   = 14303899 #from naio, so at basic prices, as intermediate products, as well
ratio= GDP27basicPrices/GDP27currentPrices
```

```python
GDPbasicPrices=[]
for i in range(28):
    GDPbasicPrices.append(GDPcurrentPrices[i]*ratio)

print(GDPbasicPrices)
```

```python
sharesGDPbasicPrices=[0]*len(GDPbasicPrices)
for i in range(len(GDPbasicPrices)):
    sharesGDPbasicPrices[i]=GDPbasicPrices[i]/14257807
sharesGDPbasicPrices
```
