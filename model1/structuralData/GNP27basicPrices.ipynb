```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
countryNames=['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 
              'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 
              'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 
              'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 
              'Slovenia', 'Spain', 'Sweden', 'totEU27']

comparison_df = pd.DataFrame()
comparison_df['country'] = countryNames
```

```python
#year 2022 GNP current prices (with VAT)
#from https://ec.europa.eu/eurostat/databrowser/view/tec00001__custom_19176318/default/table
GDPcurrentPrices=[449382.2, 561309.1, 86079.5, 67609.4, 29645.4, 286976.8, 
                  380567.4, 36300.9, 266135.0, 2653997.2, 3989390.0, 207008.7,
                  168546.2, 520718.4, 1998072.6, 36088.7, 67081.1, 76731.2,
                  17985.1, 993820.0, 661712.3, 243957.1, 280777.4, 109959.8,
                  56881.6, 1375863.0, 547190.4, 16169786.6]
#comparison_df['GDP current prices with VAT'] = GDPcurrentPrices
```

```python
#in test_ff_multi 
GDP27currentPrices = 16144780 #difference due to update values; contains VAT
GDP27basicPrices   = 14303899 #from naio, so at basic prices, as intermediate products, as well -- without VAT
ratio= GDP27basicPrices/GDP27currentPrices
```

```python
ratio
```

### GDPbasicPricesBySimulation with no adjustingCoefficient based on productivity

```python
GDPbasicPrices=[]
for i in range(28):
    GDPbasicPrices.append(GDPcurrentPrices[i]*ratio)

comparison_df['GDP basic prices'] = GDPbasicPrices
print(GDPbasicPrices)
```

```python
GDPbasicPricesBySimulation =\
[334923, 380998, 178981, 99642, 44856, 506194, 298892, 65519, 
 250589, 1700545, 2991205, 306787, 314588, 221886, 1556951, 
 90205, 157354, 40499, 58191, 758925, 1022063, 449591, 549328, 
 201911, 57885, 1501142, 387433, 14527096]
# here above values 24.03 after computing the model in test_ff_multi with no adjustingCoefficient
comparison_df['sim no adj. coeff.'] = GDPbasicPricesBySimulation
```

```python
x = np.arange(len(countryNames[0:28]))
width = 0.35
fig, ax = plt.subplots()
ax.bar(x - width/2, GDPbasicPrices[0:28], width, label='GDPbasicPrices')
ax.bar(x + width/2, GDPbasicPricesBySimulation[0:28], width, label='GDPbasicPricesBySimulation')
ax.set_ylabel("GDP")
ax.set_xticks(x)
ax.set_xticklabels(countryNames[0:28],rotation=75)
plt.legend()
plt.tight_layout()
```
