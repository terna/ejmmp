```python
mini_naio['Consumption good share'] = mini_naio['Final consumption expenditure by households']/\
    (mini_naio['Final consumption expenditure by households']+ mini_naio['Gross fixed capital formation']\
    +mini_naio['Total intermediate goods'])

mini_naio['Investment good share'] = mini_naio['Gross fixed capital formation']/\
    (mini_naio['Final consumption expenditure by households']+ mini_naio['Gross fixed capital formation']\
    +mini_naio['Total intermediate goods'])

mini_naio['Intermediate good share'] = mini_naio['Total intermediate goods']/\
    (mini_naio['Final consumption expenditure by households']+ mini_naio['Gross fixed capital formation']\
    +mini_naio['Total intermediate goods'])

pd.set_option('display.max_colwidth', 100)
mini_naio
```

```python
workforce = sbs['Total'].sum()
agriculture = 19000000 # we assume all firms of agriculture having only one worker (reflects eurostat data ~ 10 MLN) 
                       # but we also consider further 9MLN of employees (cfr. ./agricoltura.txt)
workforce +=  agriculture
agricultureFirms = agriculture / 1.9
# we are not including PA and liberi professionisti

numberOfFirms = sbs['From 0 to 9 persons employed'].sum() / 5\
              + sbs['From 10 to 19 persons employed'].sum() / 15\
              + sbs['From 20 to 49 persons employed'].sum() / 35\
              + sbs['From 50 to 249 persons employed'].sum() / 150\
              + sbs['250 persons employed or more'].sum() / 1000\
              + agricultureFirms # to be placed after in 0-9 employed class
numberOfFirms = int(numberOfFirms)
numberOfFirms
```

```python
column_names = ['#','NACE definition', 'type', 'dimensional class', 'Share of firms']  # type = consumption / investment / intermediate good
ff = pd.DataFrame(columns=column_names)

agric_tot = (mini_naio['Added value, gross'].loc[1] \
               + mini_naio['Added value, gross'].loc[2]\
               + mini_naio['Added value, gross'].loc[3])


# agriculture
share = (agricultureFirms/numberOfFirms) * mini_naio['Added value, gross'].loc[1]/ agric_tot \
                                    * mini_naio['Consumption good share'].loc[1]
ff.loc[0] = [1, "Agriculture", "C", "0-9", share]

share = (agricultureFirms/numberOfFirms) * mini_naio['Added value, gross'].loc[1]/ agric_tot \
                                    * mini_naio['Investment good share'].loc[1]
ff.loc[1] = [1,"Agriculture", "I", "0-9", share]

share = (agricultureFirms/numberOfFirms) * mini_naio['Added value, gross'].loc[1]/ agric_tot \
                                    * mini_naio['Intermediate good share'].loc[1]
ff.loc[2] = [1, "Agriculture", "Int", "0-9", share]

# silviculture
share = (agricultureFirms/numberOfFirms) * mini_naio['Added value, gross'].loc[2]/ agric_tot \
                                    * mini_naio['Consumption good share'].loc[2]
ff.loc[3] = [2, "Silviculture", "C", "0-9", share]

share = (agriculture/numberOfFirms) * mini_naio['Added value, gross'].loc[2]/ agric_tot \
                                    * mini_naio['Investment good share'].loc[2]
ff.loc[4] = [2, "Silviculture", "I", "0-9", share]

share = (agriculture/numberOfFirms) * mini_naio['Added value, gross'].loc[2]/ agric_tot \
                                    * mini_naio['Intermediate good share'].loc[2]
ff.loc[5] = [2, "Silviculture", "Int", "0-9", share]

# fishing (no investments, but with the row anyway)
share = (agriculture/numberOfFirms) * mini_naio['Added value, gross'].loc[3]/ agric_tot\
                                    * mini_naio['Consumption good share'].loc[3]              
ff.loc[6] = [3, "Fishing", "C", "0-9", share]

share = (agriculture/numberOfFirms) * mini_naio['Added value, gross'].loc[3]/ agric_tot\
                                    * mini_naio['Investment good share'].loc[3]              
ff.loc[7] = [3, "Fishing", "I", "0-9", share]

share = (agriculture/numberOfFirms) * mini_naio['Added value, gross'].loc[3]/ agric_tot\
                                    * mini_naio['Intermediate good share'].loc[2]               
ff.loc[8] = [3, "Fishing", "Int", "0-9", share]
```

In the following cell, we compute  the share of firms per each row of the firm-feature file that we are generating (each row describes the specification of the NACE sector, the type of good it produces &mdash; consumption, investment, intermediate &mdash; and its dimensional class in terms of number of employees).
We calculate the share as the ratio between number of firms of the sector (estimated through the number of workers per dimensional class in standard cases, with some exception, e.g. agriculture) and the total number of firms; whereas we decompose consumption-, investment-, and intermediate goods by using the added values.

```python
def shareCalculation(ffRow,miniNaioRow, naceDef,sbsRow):
    global numberOfFirms

    share = (sbs['From 0 to 9 persons employed'][sbsRow] / 5) / numberOfFirms * mini_naio['Consumption good share'].loc[miniNaioRow]
    ff.loc[ffRow] = [miniNaioRow, naceDef, "C", "0-9", share]

    share = (sbs['From 10 to 19 persons employed'][sbsRow] / 15) / numberOfFirms * mini_naio['Consumption good share'].loc[miniNaioRow]
    ff.loc[ffRow+1] = [miniNaioRow, naceDef, "C", "10-19", share]

    share = (sbs['From 20 to 49 persons employed'][sbsRow] / 35) / numberOfFirms * mini_naio['Consumption good share'].loc[miniNaioRow]
    ff.loc[ffRow+2] = [miniNaioRow, naceDef, "C", "20-49", share]

    share = (sbs['From 50 to 249 persons employed'][sbsRow] / 150) / numberOfFirms * mini_naio['Consumption good share'].loc[miniNaioRow]
    ff.loc[ffRow+3] = [miniNaioRow, naceDef, "C", "50-249", share]

    share = (sbs['250 persons employed or more'][sbsRow] / 1000) / numberOfFirms * mini_naio['Consumption good share'].loc[miniNaioRow]
    ff.loc[ffRow+4] = [miniNaioRow, naceDef, "C", ">=250", share]

    share = (sbs['From 0 to 9 persons employed'][sbsRow] / 5) / numberOfFirms * mini_naio['Investment good share'].loc[miniNaioRow]
    ff.loc[ffRow+5] = [miniNaioRow, naceDef, "I", "0-9", share]
    
    share = (sbs['From 10 to 19 persons employed'][sbsRow] / 15) / numberOfFirms * mini_naio['Investment good share'].loc[miniNaioRow]
    ff.loc[ffRow+6] = [miniNaioRow, naceDef, "I", "10-19", share]

    share = (sbs['From 20 to 49 persons employed'][sbsRow] / 35) / numberOfFirms * mini_naio['Investment good share'].loc[miniNaioRow]
    ff.loc[ffRow+7] = [miniNaioRow, naceDef, "I", "20-49", share]

    share = (sbs['From 50 to 249 persons employed'][sbsRow] / 150) / numberOfFirms * mini_naio['Investment good share'].loc[miniNaioRow]
    ff.loc[ffRow+8] = [miniNaioRow, naceDef, "I", "50-249", share]

    share = (sbs['250 persons employed or more'][sbsRow] / 1000) / numberOfFirms * mini_naio['Investment good share'].loc[miniNaioRow]
    ff.loc[ffRow+9] = [miniNaioRow, naceDef, "I", ">=250", share]

    share = (sbs['From 0 to 9 persons employed'][sbsRow] / 5) / numberOfFirms * mini_naio['Intermediate good share'].loc[miniNaioRow]
    ff.loc[ffRow+10] = [miniNaioRow, naceDef, "Int", "0-9", share]
    
    share = (sbs['From 10 to 19 persons employed'][sbsRow] / 15) / numberOfFirms * mini_naio['Intermediate good share'].loc[miniNaioRow]
    ff.loc[ffRow+11] = [miniNaioRow, naceDef, "Int", "10-19", share]

    share = (sbs['From 20 to 49 persons employed'][sbsRow] / 35) / numberOfFirms * mini_naio['Intermediate good share'].loc[miniNaioRow]
    ff.loc[ffRow+12] = [miniNaioRow, naceDef, "Int", "20-49", share]

    share = (sbs['From 50 to 249 persons employed'][sbsRow] / 150) / numberOfFirms * mini_naio['Intermediate good share'].loc[miniNaioRow]
    ff.loc[ffRow+13] = [miniNaioRow, naceDef, "Int", "50-249", share]

    share = (sbs['250 persons employed or more'][sbsRow] / 1000) / numberOfFirms * mini_naio['Intermediate good share'].loc[miniNaioRow]
    ff.loc[ffRow+14] = [miniNaioRow, naceDef, "Int", ">=250", share]
    
```
