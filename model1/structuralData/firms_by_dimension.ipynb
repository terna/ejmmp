**Linear case**

With $m$ mean value:  
$\Delta b = \frac{2\,(b_1 - m)}{n-1}, \qquad b_1 > m$

```python
# prova
b_1 =  6500000
S   =  30393757
n   = 9
m   = S/n
D_b = 2 * (b_1 - m)/(n-1)
#b_n = 2*S/n - b_1

b_n = b_1 - (n-1) * D_b
b_n

b_vector = []

for i in range(0,n):
    b_vector.append(b_1 - D_b*i)
    print(i, b_1 - D_b*i)

total = sum(b_vector)
total
```
