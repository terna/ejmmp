```python
#sbs_austria = pd.read_excel('sbs_austria.xlsx', sheet_name = 'Sheet 1')
eu27_list = ['European Union - 27 countries (from 2020)', 
             'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 
            'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 
            'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 
            'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 
            'Slovenia', 'Spain', 'Sweden']

addedvalue_df = pd.read_excel('addedvalue_by_country.xlsx', sheet_name = 'Sheet 1')


addedvalue_df.columns = addedvalue_df.iloc[9]

addedvalue_df = addedvalue_df.iloc[11:39,:]
addedvalue_df = addedvalue_df.reset_index(drop=True)
addedvalue_df = addedvalue_df.loc[:, addedvalue_df.columns.notna()]
addedvalue_df = addedvalue_df.set_index(addedvalue_df.columns[0])
addedvalue_df = addedvalue_df.reindex(eu27_list)
addedvalue_df = addedvalue_df.replace(":", np.nan)

#from this file we will use 2020 as a reference year to get the share of each country in terms of added value
#we use 2020 as it is the most complete information (all countries but Bulgary) and then we get the Bulgary value by difference



addedvalue_df.iloc[3,5] = addedvalue_df['2020'].iloc[0]- addedvalue_df['2020'].iloc[1:].sum()
addedvalue_df = addedvalue_df.iloc[1:] 

addedvalue_df['Share in 2020'] = (addedvalue_df['2020'] / addedvalue_df['2020'].sum() )

addedvalue_countryshares = addedvalue_df['Share in 2020'].to_frame()
addedvalue_countryshares.index.name = "Country"
addedvalue_countryshares
```
