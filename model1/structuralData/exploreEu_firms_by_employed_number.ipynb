```python
import pickle
import numpy as np
```

```python
with open('eu_firms_by_employed_number.xp', 'rb') as f:
            eu_firms_by_employed_number = pickle.load(f)
```

```python
import pandas as pd

results = []

for i, df in enumerate(eu_firms_by_employed_number):

    # totale per riga
    row_totals = df.sum(axis=1)

    # differenza rispetto a 1
    diff_from_one = row_totals - 1

    # se il totale è 0 → metti 0
    diff_from_one = diff_from_one.where(row_totals != 0, 0)

    summary = pd.DataFrame({
        "row_total": row_totals,
        "diff_from_1": diff_from_one
    })

    # filtro: totale != 0 e errore > 1e-12
    mask = (row_totals != 0) & (diff_from_one.abs() > 0)

    print(f"\nDataFrame {i}")
    display(summary[mask])

    results.append(summary)
```
