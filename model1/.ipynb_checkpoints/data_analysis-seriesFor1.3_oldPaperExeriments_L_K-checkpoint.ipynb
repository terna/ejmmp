```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os
os.chdir("./oldPaperExperiments")
myListDir=os.listdir()

folder=input("The name of your input folder without extension: ")
while not folder+".0" in myListDir: #extension 0 certainly exists, as we have at least one rank
    print(folder+": wrong name")
    folder=input("The name of your input folder, without extension: ")

os.chdir("./"+folder+".0") #extension 0 certainly exists, as we have at least one rank

with open("plotInfo.csv","r") as data:
    reader = csv.reader(data, delimiter = ',')
    for row in reader:
        fileNameRoot=row[0] #the names of the files contating the data start with...
        rankNum=int(row[1]) #we have file from different ranks, reported in different folders with name, for a specific
                            #rank: fileNameRoot+"."+str(rank)
        #ag0num=int(row[2])  #how many agents in each rank, with type==0

#current folders
print("Current folders are:")
for k in range(rankNum):
    print(folder+"."+str(k))
```

```python
names=["_updatedLabor_","_updatedCapital_"]
```

```python
# series database
series = {}

for name in names:
    k = 0
    os.chdir("../" + folder + "." + str(k))

    df = pd.read_csv(fileNameRoot + name + str(k) + ".csv", header=None)
    df.columns = range(1, df.shape[1] + 1)   # 1..8 (o comunque 1..n)
    series[name] = df

    for k in range(1, rankNum):
        os.chdir("../" + folder + "." + str(k))

        df_k = pd.read_csv(fileNameRoot + name + str(k) + ".csv", header=None)
        df_k.columns = series[name].columns  # garantisce stesse colonne
        series[name] = series[name].add(df_k)
```

```python
for name in names:
    display(name)
    display(series[name])
    display("____________________________________________________________")
    series[name].plot(title="Legenda: firm class types",xlabel="time",ylabel=name)
```

```python
# Plot each series, move legend to the far right, and save figures in ./newPlots
from pathlib import Path
import matplotlib.pyplot as plt

out_dir = Path.cwd() / "newPlots"
out_dir.mkdir(parents=True, exist_ok=True)

for name in names:
    ax = series[name].plot(
        title="Legenda: firm class types",
        xlabel="time",
        ylabel=name,
        legend=True
    )

    # Put the legend outside on the right
    leg = ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), borderaxespad=0.0)
    fig = ax.get_figure()

    # Leave room for the legend on the right
    fig.tight_layout()

    # Safe filename
    safe = "".join(ch if ch.isalnum() or ch in ("-", "_", ".") else "_" for ch in str(name))
    fig.savefig(out_dir / f"{safe}.png", bbox_inches="tight", dpi=200)

    plt.close(fig)

print(f"Saved {len(names)} plot(s) into: {out_dir.resolve()}")
```
