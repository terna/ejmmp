```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# reference year: 2022
# https://ec.europa.eu/eurostat/databrowser/view/nama_10_a10__custom_20687844/default/table?lang=en
# https://ec.europa.eu/eurostat/databrowser/view/nama_10_a10_e/default/table?lang=en

av_agri = pd.read_excel("nama_av_agri.xlsx", sheet_name = "Sheet 1")
av_tot = pd.read_excel("nama_tot_av.xlsx" , sheet_name = "Sheet 1")
tot_emp = pd.read_excel("nama_tot_emp.xlsx" , sheet_name = "Sheet 1")
emp = pd.read_excel("nama_emp.xlsx" , sheet_name = "Sheet 1")
agri_tot_emp = pd.read_excel("agri_tot_emp.xlsx" , sheet_name = "Sheet 1")
agri_emp = pd.read_excel("agri_emp.xlsx" , sheet_name = "Sheet 1")

nama_av_files = [av_agri, av_tot, tot_emp, emp, agri_tot_emp, agri_emp]

countryNames=['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 
              'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 
              'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 
              'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 
              'Slovenia', 'Spain', 'Sweden', 'totEU27']


for i in range(len(nama_av_files)):
    nama_av_files[i] = nama_av_files[i][10:38]
    nama_av_files[i] = nama_av_files[i].drop("Unnamed: 2", axis=1)
    nama_av_files[i].columns = ['country', 'AV']
    nama_av_files[i].iloc[0,0] = 'totEU27'
    nama_av_files[i]["country"] = pd.Categorical(nama_av_files[i]["country"], categories=countryNames, ordered=True)
    nama_av_files[i] = nama_av_files[i].sort_values("country")
    nama_av_files[i] = nama_av_files[i].reset_index(drop=True)
```

```python
nama_av = pd.DataFrame()
nama_av['country'] = nama_av_files[0]['country']
nama_av['agri_AV'] = nama_av_files[0]['AV']
nama_av['all_nace_AV'] =  nama_av_files[1]['AV']

nama_av['nace_noagri_AV'] = nama_av['all_nace_AV'] - nama_av['agri_AV'] 

nama_av['tot_emp'] = nama_av_files[2]['AV']
nama_av['emp'] = nama_av_files[3]['AV']
nama_av['agri_tot_emp'] = nama_av_files[4]['AV']
nama_av['agri_emp'] = nama_av_files[5]['AV']

nama_av['nace_noagri_tot_emp'] = nama_av['tot_emp'] - nama_av['agri_tot_emp'] 
nama_av['nace_noagri_emp'] = nama_av['emp'] - nama_av['agri_emp'] 

nama_av['tot_prod'] = nama_av['all_nace_AV'] / nama_av['tot_emp'] 
nama_av['prod'] = nama_av['all_nace_AV'] / nama_av['emp'] 

nama_av['agri_tot_prod'] = nama_av['agri_AV']/ nama_av['agri_tot_emp'] 
nama_av['agri_prod'] = nama_av['agri_AV']/ nama_av['agri_emp']

nama_av['noagri_tot_prod'] = nama_av['nace_noagri_AV'] / nama_av['nace_noagri_tot_emp']
nama_av['noagri_prod'] = nama_av['nace_noagri_AV'] / nama_av['nace_noagri_emp']


# NB: tot_prod accounts also for self-employers, prod accounts only for the employees

nama_av
```

```python
eu_productivities = pd.DataFrame()
eu_productivities['country'] = nama_av['country']

eu_productivities['agri tot prod'] = nama_av['agri_tot_prod'] / nama_av.iloc[-1, 12] #  totEU27, 'agri_tot_prod'
eu_productivities['no agri tot prod'] = nama_av['noagri_tot_prod'] /  nama_av.iloc[-1, 14]  # totEU27, 'noagri_tot_prod'

eu_productivities['no agri tot prod Italy-adjusted'] = nama_av['noagri_tot_prod'] /  nama_av.iloc[14, 14] 
eu_productivities
```

```python
eu_productivities.to_pickle("./eu_productivities.xp")
nama_av.to_pickle("./nama_av.xp")
```
