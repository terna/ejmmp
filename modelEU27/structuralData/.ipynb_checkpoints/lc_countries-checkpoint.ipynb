```python
import pandas as pd
```

```python
# data available at https://ec.europa.eu/eurostat/statistics-explained/index.php?title=Hourly_labour_costs
# https://ec.europa.eu/eurostat/statistics-explained/images/2/22/Hourly_labour_costs_in_€_2024_rev1.png


countries = [
    "totEU27", "Belgium", "Bulgaria", "Czechia", "Denmark", "Germany",
    "Estonia", "Ireland", "Greece", "Spain", "France", "Croatia",
    "Italy", "Cyprus", "Latvia", "Lithuania", "Luxembourg", "Hungary",
    "Malta", "Netherlands", "Austria", "Poland", "Portugal", "Romania",
    "Slovenia", "Slovakia", "Finland", "Sweden"
]



values = [
    28.4, 40.5, 6.6, 14.6, 44.9, 36.8, 14.1, 33.5, 13.8,
    23.0, 38.8, 11.3, 29.2, 17.3, 10.7, 10.3, 47.3, 9.8, 15.6,
    37.7, 36.4, 10.9, 15.1, 8.2, 20.7, 13.7, 34.0, 38.0
]

lc_countries = pd.DataFrame({
    "Country": countries,
    "lc": values
})

countryNames=['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 
              'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 
              'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 
              'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 
              'Slovenia', 'Spain', 'Sweden', 'totEU27']
lc_countries = lc_countries.set_index("Country").reindex(countryNames).reset_index()

lc_countries['index'] = lc_countries['lc']/28.4

lc_countries
```

```python
lc_countries['index'].to_pickle("./lc_countries.xp")
```
