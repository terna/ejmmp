```python
import csv
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_rows', 1000)
```

```python
tec01gdp = pd.read_excel('tec00001__custom_21446525_page_spreadsheet.xlsx', sheet_name = 'Sheet 1')
tec01gdp = tec01gdp.iloc[9:37, 0:2]
tec01gdp = tec01gdp.reset_index(drop=True)

tec01gov_consumption =  pd.read_excel('tec00010__custom_21446921_page_spreadsheet.xlsx', sheet_name = 'Sheet 1')
tec01gov_consumption = tec01gov_consumption.iloc[9:37, 0:2]
tec01gov_consumption = tec01gov_consumption.reset_index(drop=True)

tec01private_consumption =  pd.read_excel('tec00009__custom_21447046_page_spreadsheet.xlsx', sheet_name = 'Sheet 1')
tec01private_consumption =  tec01private_consumption.iloc[9:37, 0:2]
tec01private_consumption =  tec01private_consumption.reset_index(drop=True)

tec01 = tec01gdp 
tec01['private consumption'] = tec01private_consumption['Unnamed: 1']
tec01['gov consumption'] = tec01gov_consumption['Unnamed: 1']
tec01 = tec01.rename(columns={"Data extracted on 13/05/2026 16:54:17 from [ESTAT]": "country", "Unnamed: 1": "gdp"})

tec01.iloc[0,0] = 'EU27'

tec01

tec01['private_cons / gdp'] = tec01['private consumption']/tec01['gdp']
tec01['gov_cons /gdp'] = tec01['gov consumption']/tec01['gdp']

eu_countries = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 
                             'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 
                             'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 
                             'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 
                             'Slovenia', 'Spain', 'Sweden', 'totEU27']


tec01["country"] = tec01["country"].replace({"EU27": "totEU27"})

tec01 = (tec01.set_index("country").reindex(eu_countries))



tec01
```
